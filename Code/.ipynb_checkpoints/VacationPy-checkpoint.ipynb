```python
import requests
import pandas as pd
import numpy as np
import gmaps
from config import gkey
import json
from pprint import pprint

gmaps.configure(api_key = gkey)
```

```python
path = "../Output_Data/Weather_data.csv"

weather_df = pd.DataFrame(pd.read_csv(path))

weather_df
```

```python
lat_lng = []
humid = []

for index, row in weather_df.iterrows():
    
    if row["Humidity"] >= 85:
    
        lat_lng.append((row["Lat"], row["Lng"]))
                                        
# Plot Heatmap
figure_layout = {
    'width': '1000px',
    'height': '750px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)

fig.add_layer(gmaps.heatmap_layer(lat_lng))
fig
```

```python
weather_df = weather_df.loc[weather_df['Max Temp'] < 80]
weather_df = weather_df.loc[weather_df['Max Temp'] > 70]
weather_df = weather_df.loc[weather_df['Wind Speed'] < 10]
weather_df = weather_df.loc[weather_df['Cloudiness'] == 0]

weather_df.head()
```

```python
geocords = []

for index, row in weather_df.iterrows():
    
    geocords.append((row["Lat"], row["Lng"]))
                                        
print(geocords)
```

```python
# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
hotels = []
city = []
country = []

for cord in geocords:
    
    # Parameters dictionary
    params = {
        "location": cord,
        "keyword": "hotel",
        "radius": 5000,
        "type": "hotel",
        "key": gkey
    }

    # run a request using our params dictionary
    response = requests.get(base_url, params=params).json()

    hotels.append(response['results'][0]['name'])
    print(hotels)

#     pprint(json.dumps(response,indent=4,sort_keys=True))
```

```python
# Plot Heatmap
figure_layout = {
    'width': '1000px',
    'height': '750px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)

fig.add_layer(gmaps.heatmap_layer(lat_lng))


# Assign the marker layer to a variable
markers = gmaps.marker_layer(geocords)
# Add the layer to the map
fig.add_layer(markers)
fig
```
